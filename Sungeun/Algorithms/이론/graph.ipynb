## 1차원 그래프 vs 2차원 그리드 차이점
#### 최단 거리 배열 (distance)

1차원: distance = [INF] * (N + 1) (1차원 리스트)

**2차원: distance = [[INF] * N for _ in range(N)] (2차원 리스트)**

#### 우선순위 큐에 들어가는 데이터

1차원: (누적 비용, 노드 번호)

**2차원: (누적 비용, r좌표, c좌표)**

#### 연결된 노드(간선) 탐색 방식

1차원: 만들어둔 graph[now] 리스트를 for문으로 순회

**2차원: 현재 좌표 상하좌우 4방향(dr, dc)을 계산하여 그리드 범위 내에 있는지 확인**

# kruskal algorithms

### 크루스칼(Kruskal) 알고리즘의 핵심 논리 3단계
크루스칼은 그래프 내의 모든 노드를 최소 비용으로 연결하는 **최소 신장 트리(MST)**를 찾는 알고리즘입니다. 구조는 다익스트라보다 오히려 훨씬 단순하고 직관적입니다.

정렬: 모든 간선(Edge)을 '비용(cost)'이 적은 순서대로 오름차순 정렬합니다.

선택 및 사이클 확인 (Find): 비용이 가장 적은 간선부터 하나씩 확인하며, 이 간선을 연결했을 때 사이클(순환)이 생기는지 확인합니다. (이때 find 함수를 씁니다.)

연결 (Union): 사이클이 생기지 않는다면 두 노드를 연결하고(union), 비용을 누적합니다. 사이클이 생기면 그냥 버리고 다음 간선으로 넘어갑니다.

#### 1. 크루스칼 (Kruskal) : "전국구 최저가 줍기" (Edge 기준)
시야: 그래프 전체를 한 번에 조망합니다.

방식: 노드가 어디서 어떻게 연결되어 있든 상관없이, 일단 전국에 있는 모든 간선을 가격표 순으로 오름차순 정렬합니다. 그리고 가장 싼 간선부터 무조건 줍습니다.

주의점: 줍다 보니 우연히 길이 둥글게 이어져 버리면(사이클) 버립니다. (이때 유니온-파인드 사용)

비유: 중고나라에서 지역 상관없이 '전국 최저가' 순으로 검색해서 물건을 사 모으는 방식입니다.

#### 2. 프림 (Prim) : "야금야금 영토 확장" (Node 기준)
시야: 내가 현재 서 있는 '노드(내 땅)' 주변만 봅니다.

방식: 시작 노드 하나에서 출발합니다. 현재 내 땅(지금까지 방문한 노드들)과 바로 맞닿아 있는 간선들 중에서 가장 싼 것을 골라 다음 노드를 내 땅으로 편입시킵니다.

주의점: 내 땅 안에서 이미 연결된 노드끼리 또 이어지는 간선은 무시합니다. (이때 visited 배열 사용)

비유: 스타크래프트나 문명 게임에서 본진(시작 노드)을 중심으로 시야를 넓혀가며 야금야금 주변 영토를 확장해 나가는 방식입니다.
